# Random Forest

```python
import pandas as pd

# Path to your CSV file
csv_file_path = 'NDMN-Notebooks/CostaRica-Notebook/Costa_Rica_Data/Costa_Classification_Data_Cleaned2.csv'

# Load the CSV into a DataFrame
df = pd.read_csv(csv_file_path)

# Display the first few rows
df.head()
```

```python
# Check for columns with missing values (NaNs)
missing_columns = df.columns[df.isnull().any()]

# Display the results
print("Columns with missing values:", missing_columns)
```

```python
print(df.columns.tolist())
```

```python
y = df['Use']  # Set the target variable
X = df[['ndvi','evi','savi','brightness','wetness','slope','aspect','mTPI','ocs_1mMed', 'sand_1mMed','topDiv']]  # Predictor features

# Check the result
print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
X.head()
```

```python
print(y.unique())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=30, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

```python
y_prediction = rf.predict(X_test)
```

```python
rf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_prediction))
```

# One Hot Encoder

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')
```

```python
categorical_features = ['Vegetation','Cover', 'Terrain']
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')

encoded_features = ohe.fit_transform(df[categorical_features])
df = pd.concat([df.drop(columns=categorical_features), encoded_features], axis=1)
```

```python
y = df['Use']  # Set the target variable
X = df.drop(columns=['Use'])  # Drop the target column to get the features

X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=30, test_size=0.2)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_prediction = rf.predict(X_test)
```

```python
rf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_prediction))
```

```python
features = pd.DataFrame(rf.feature_importances_, index=X.columns, columns=['Importance'])
features = features.sort_values(by='Importance', ascending=False)
print(features)
# Most important features in the model
```
